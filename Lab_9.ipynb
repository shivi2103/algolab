# Shortest Path Using Dijkstra Algorithm in Python
Problem Description:
Given a weighted graph with non negative edge weights. Find the shortest distance from a source vertex to all other vertices using Dijkstra algorithm

#Theory
Dijkstra algorithm finds shortest distance from one source node to all other nodes in a weighted graph with non negative weights. Dictionary structure stores nodes as keys and neighbor nodes with weights as values.

##Algorithm Steps

1. Represent graph using dictionary.
2. Set distance of every node as infinity.
3. Set distance of source node as 0.
4. Select node with smallest distance among unvisited nodes.
5. Update distance of neighbor nodes.
6. Repeat until all nodes get visited.

```python
def dijkstra(graph, start):

    distances = {}
    visited = []

    for node in graph:
        distances[node] = float('inf')

    distances[start] = 0

    while len(visited) < len(graph):

        min_node = None
        min_distance = float('inf')

        for node in graph:
            if node not in visited and distances[node] < min_distance:
                min_distance = distances[node]
                min_node = node

        visited.append(min_node)

        for neighbor, weight in graph[min_node].items():
            new_distance = distances[min_node] + weight

            if new_distance < distances[neighbor]:
                distances[neighbor] = new_distance

    return distances
```

```python
graph = {
'A': {'B': 4, 'C': 2},
'B': {'A': 4, 'C': 5, 'D': 10},
'C': {'A': 2, 'B': 5, 'D': 3},
'D': {'B': 10, 'C': 3}
}

result = dijkstra(graph, 'A')

for node in result:
    print("Distance from A to", node, "=", result[node])
```

1. Modify program for user input graph using dictionary.
2. Display shortest path with total distance.
3. Add two more nodes in graph.
4. Test algorithm with different source nodes.

```python
import heapq
def build_graph():
    graph = {}
    n = int(input("Enter number of nodes: "))
    for _ in range(n):
        node = input("Enter node name: ").strip()
        graph[node] = {}
        m = int(input(f"Enter number of neighbors for {node}: "))
        for _ in range(m):
            neighbor = input(f"Enter neighbor of {node}: ").strip()
            weight = int(input(f"Enter weight from {node} to {neighbor}: "))
            graph[node][neighbor] = weight
    return graph

# 2
def dijkstra(graph, start):
    distances = {node: float('inf') for node in graph}
    distances[start] = 0
    previous_nodes = {node: None for node in graph}
    pq = [(0, start)]

    while pq:
        current_distance, current_node = heapq.heappop(pq)

        if current_distance > distances[current_node]:
            continue

        for neighbor, weight in graph[current_node].items():
            distance = current_distance + weight
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                previous_nodes[neighbor] = current_node
                heapq.heappush(pq, (distance, neighbor))

    return distances, previous_nodes

# 3
def shortest_path(previous_nodes, start, target):
    path = []
    current = target
    while current is not None:
        path.insert(0, current)
        current = previous_nodes[current]
    return path if path[0] == start else []

# 4
graph = build_graph()
graph['E'] = {'D': 3, 'F': 2}
graph['F'] = {'E': 2, 'C': 7}

print("\nGraph after adding nodes E and F:")
print(graph)
while True:
    source = input("\nEnter source node (or 'exit' to quit): ").strip()
    if source.lower() == 'exit':
        break
    target = input("Enter target node: ").strip()

    if source not in graph or target not in graph:
        print("Invalid nodes. Try again.")
        continue

    distances, previous_nodes = dijkstra(graph, source)
    path = shortest_path(previous_nodes, source, target)

    if path:
        print(f"Shortest path from {source} to {target}: {' -> '.join(path)}")
        print(f"Total distance: {distances[target]}")
    else:
        print("No path found.")
```
